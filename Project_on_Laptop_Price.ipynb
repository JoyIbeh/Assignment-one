#DATA VISUALIZATION

1.
Which brand has the highest average price among the laptops?

```python
# Group the dataset by brand and calculate the mean of price
lp1_brd = lp1.groupby("brand").mean().sort_values(by= "Price", ascending= False)
lp1_brd
# Find the brand with the highest average price
max_price_brand = lp1_brd.idxmax()
max_price_value = lp1_brd.max()

print (lp1_brd)
print ('Brand with the highest average price is', max_price_brand)
print ('The Value of the Brand with the highest average price is', max_price_value)
```

```python
# Plot a bar chart of the average price by brand
lp1_brand = lp1.groupby("brand").mean()["Price"]
lp1_brand

plt.figure(figsize=(6, 6))
plt.bar(lp1_brand.index, lp1_brand, color="Blue")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.title("Average Price of Laptops by Brand")
plt.show()
```

4. Which operating system and bit version are most common among the laptops?

```python
# Group the dataframe by the os and os_bit columns and count the number of laptops in each group
lp1_os = lp1.groupby(["os", "os_bit"]).size().reset_index(name="count")
# Print the os and os_bit combination that has the most laptops
max_os = lp1_os.loc[lp1_os["count"].idxmax()]
print("The most common operating system and bit version among the laptops is:", max_os["os"], max_os["os_bit"])
```
